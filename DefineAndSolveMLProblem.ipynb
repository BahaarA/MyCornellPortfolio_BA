```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

<b>Task:</b> Use the rest of this notebook to carry out your project plan. 

You will:

1. Prepare your data for your model.
2. Fit your model to the training data and evaluate your model.
3. Improve your model's performance by performing model selection and/or feature selection techniques to find best model for your problem.

Add code cells below and populate the notebook with commentary, code, analyses, results, and figures as you see fit. 

```python
# Data Preparation
df_encoded = pd.get_dummies(df, drop_first=True)

# Dropping irrelevant feature
df_encoded.drop(['fnlwgt'], axis=1, inplace=True)

# Checking column names after encoding
print("Columns in df_encoded:", df_encoded.columns)

# Handling missing values by filling only numeric columns
numeric_columns = df_encoded.select_dtypes(include=['number']).columns
df_encoded[numeric_columns] = df_encoded[numeric_columns].fillna(df_encoded[numeric_columns].median())

# Scaling numerical features
scaler = StandardScaler()
df_encoded[['age', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']] = scaler.fit_transform(
    df_encoded[['age', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']]
)

# Verify column names again and ensure 'income_binary' is correctly named
print("Columns in df_encoded after scaling:", df_encoded.columns)

# Fixing the issue: update 'income_binary' to the correct column name if needed
target_column = 'income_binary_>50K'  # Update this based on the actual column name

# Splitting the data into training and test sets
X = df_encoded.drop(target_column, axis=1)
y = df_encoded[target_column]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model Training: Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Model Evaluation: Logistic Regression
y_pred = log_reg.predict(X_test)

print("Logistic Regression Performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(classification_report(y_test, y_pred))

conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix for Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Model Improvement: Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

print("Random Forest Performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_rf)}")
print(classification_report(y_test, y_pred_rf))

conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Greens')
plt.title("Confusion Matrix for Random Forest")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
